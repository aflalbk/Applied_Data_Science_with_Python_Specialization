```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cwurData.csv')
df.head()
```

```python
def create_tire(rank):
  if 1 <= rank <= 100:
    return 'tire one'
  elif 101 <= rank <= 200:
    return 'tire two'
  elif 201 <= rank <= 300:
    return 'tire three'
  else:
    return 'other top university'

df['rank_level'] = df['world_rank'].apply(lambda x : create_tire(x))
df.head()
```

```python
df.pivot_table(index='country', columns='rank_level', values='score', aggfunc=[np.mean]).head()
```

```python
df.pivot_table(index='country', columns='rank_level', values='score', aggfunc=[np.max, np.mean]).head()
```

```python
df.pivot_table(index='country', columns='rank_level', values='score', aggfunc=[np.max, np.mean], margins=True).head()
```

```python
new_df = df.pivot_table(index='country', columns='rank_level', values='score', aggfunc=[np.max, np.mean], margins=True)

print(new_df.index)
print(new_df.columns)
```

```python
new_df['mean']['tire one'].head()
```

```python
# transpose the innermost column to row. "Stacking"

new1_df = new_df.stack()
new1_df.head()
```
