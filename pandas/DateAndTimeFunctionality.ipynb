DatetimeIndex and PeriodIndex

Converting to Datetime

```python
d1=['2 june 2023','2/3/2013','24-03-2024','20 may, 2020']

ts1 = pd.DataFrame(np.random.randint(10,100,(4,2)), index=d1, columns=list('ab'))
ts1
```

```python
ts1.index = pd.to_datetime(ts1.index)
ts1
```

```python
pd.to_datetime('4,7,2020', dayfirst=True)
```

**Working with dates in a dataframe**

```python
dates = pd.date_range('10-24-2023', periods=9, freq='2W-SUN')
dates
```

```python
df = pd.DataFrame({'count_1' : 100+np.random.randint(-5,10,9).cumsum(),
                   'count_2' : 120+np.random.randint(-5,10,9)}, index=dates)
df
```

```python
df.diff()
```

```python
df.resample('M').mean()
```
