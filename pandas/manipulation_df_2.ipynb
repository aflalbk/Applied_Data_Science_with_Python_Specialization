```python
import pandas as pd
```

```python
df = pd.read_csv("presidents.csv")
df.head()
```

```python
df = df.set_index('#')
df.head()
```

```python
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

##### Let's clean the data.

```python
# preident name into first name and last name.

df['first name'] = df['president'].str.split().str[0]
df['last name'] = df['president'].str.split().str[-1]

df.head()
```

###### another method 

```python
del(df['first name'])
del(df['last name'])
```

```python
def split_president_name(row):
    row['first name'] = row['president'].split(" ")[0]
    row['last name'] = row['president'].split(" ")[-1]
    return row


df = df.apply(split_president_name, axis='columns')
df.head()
```

###### another method

```python
del(df['first name'])
del(df['last name'])
```

```python
pattern = "(?P<first_name>^[\w]*)(?:.* )(?P<last_name>[\w]*$)"

df[['first name','last name']] = df['president'].str.extract(pattern)
df.head()
```

```python
df['born'] = df['born'].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df.head()
```
