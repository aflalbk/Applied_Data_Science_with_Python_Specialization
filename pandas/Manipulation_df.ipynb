```python
import pandas as pd
```

```python
df = pd.read_csv('log.csv')
print(len(df))
print(df.head(20))
```

```python
df = df.set_index('time')
df = df.sort_index()
df
```

```python
df = df.reset_index()
df = df.set_index(['time','user'])
df
```

```python
df.sort_index()
```

```python
df = df.fillna(method = 'ffill')
df
```

```python
# we can use replace function in this let's try

df_new = df.copy()
df_new.head()
```

```python
# replace all bob with boby 

df_new = df_new.reset_index()
df_new = df_new.replace('bob','boby')
df_new
```

```python
df_new = df_new.replace(['boby','sue'],['bob','susu'])
df_new
```
