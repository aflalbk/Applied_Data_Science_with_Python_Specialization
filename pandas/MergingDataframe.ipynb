```python
import pandas as pd
```

```python
# creat a data frame
staf_df = pd.DataFrame([{'name' : 'akil', 'role' : 'HR'},
                        {'name' : 'manu', 'role' : 'examiner'},
                        {'name' : 'john', 'role' : 'Course liasion'},
                        {'name' : 'abdu', 'role' : 'HoD'}])

staf_df = staf_df.set_index('name')
student_df = pd.DataFrame([{'name' : 'abdu', 'school' : 'Bussiness'},
                           {'name' : 'mike', 'school' : 'law'},
                           {'name' : 'manu', 'school' : 'Engineering'}])
student_df = student_df.set_index('name')

print(staf_df)
print(student_df)
```

```python
# let's do outer join of theies two columns. (union of data fram)

pd.merge(staf_df, student_df, how='outer', left_index=True, right_index=True)
```

```python
# let's do inner join of the data frame. intersection of data frame

pd.merge(staf_df, student_df, how='inner', left_index=True, right_index=True)
```

```python
# let's do Left join. it take left data frame. then join right data frame's rows with left index.

pd.merge(staf_df, student_df, how='left', left_index=True, right_index=True)
```

```python
# let's do right join. it takes right data frame, then join left data frame's rows with right index.

pd.merge(staf_df, student_df, how='right', left_index=True, right_index=True)
```

 Above method using join based on indices. but we can also join with columns. using `On` parameter.

```python
staf_df = staf_df.reset_index()
student_df = student_df.reset_index()

pd.merge(staf_df, student_df, how='right', on='name')
```
