```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# let's do hypothesis test

df = pd.read_csv('grades.csv')
df.head()
```

```python
print(f"number of row: {df.shape[0]} and number of columns: {df.shape[1]}")
```

```python
# let's segment this population
# 1 - 'early finish' = before or 2015-12-30
# 2 - 'late finish' = after the date

early_fnsh = df[pd.to_datetime(df['assignment1_submission'])<'2016']
early_fnsh.head()
```

```python
late_fnsh = df[pd.to_datetime(df['assignment1_submission'])>='2016']
late_fnsh.head()
```

```python
late_fnsh = df[~df.index.isin(early_fnsh.index)]
late_fnsh.head()
```

```python
print(early_fnsh.shape[0])
print(late_fnsh.shape[0])
print(early_fnsh.shape[0]+late_fnsh.shape[0])
```

```python
print(f"mean assignment1_grade of early finish: {early_fnsh['assignment1_grade'].mean()}")
print(f"mean assignment1_grade of late finish : {late_fnsh['assignment1_grade'].mean()}")
```

its looks similar but are they the same?

hear the t-test comes in

alternative hypothesis ("these are different") and null hypothesis ("these are the same")

alpha = 5% or 0.05

we use **`ttest_ind()`** function

```python
from scipy.stats import ttest_ind

ttest_ind(early_fnsh['assignment1_grade'],late_fnsh['assignment1_grade'])
```

it is 0.18, which is > 0.05. so we can't reject null hypothesis.

that means we don't have enough certainty in our evidency to come to a conclusion to the contrary.

```python
# let's try in all other assignments.
print(ttest_ind(early_fnsh['assignment2_grade'], late_fnsh['assignment2_grade']))
print(ttest_ind(early_fnsh['assignment3_grade'], late_fnsh['assignment3_grade']))
print(ttest_ind(early_fnsh['assignment4_grade'], late_fnsh['assignment4_grade']))
print(ttest_ind(early_fnsh['assignment5_grade'], late_fnsh['assignment5_grade']))
print(ttest_ind(early_fnsh['assignment6_grade'], late_fnsh['assignment6_grade']))
```

```python
df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.shape
```

```python
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.shape
```

```python
# we do t-test in every columns in both data frame

def test_columns(alpha=0.1):
  num_diff = 0
  for col in df1.columns:
    teststat= ttest_ind(df1[col],df2[col])
    pval = teststat.pvalue
    if pval <= alpha:
      print(f"column {col} is satisticaly significantly different at alpha={alpha}, pvalue={pval}")
      num_diff += 1

  print("total num of diff was {} which is {}% ".format(num_diff, num_diff/len(df1.columns)*100))

test_columns()
```
