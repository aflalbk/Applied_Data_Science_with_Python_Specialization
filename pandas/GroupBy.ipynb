```python
from pandas.core.indexing import convert_missing_indexer
import pandas as pd
import numpy as np

df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
df.head()
```

```python
%%timeit -n 10

for state in df['STNAME'].unique():
  avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
  print(f'{avg} is the average city population of {state}')
```

```python
%%timeit -n 3

for group, frame in df.groupby('STNAME'):
  avg = np.average(frame['CENSUS2010POP'])
  print(f'{avg} is the avg city population of {group}')
```

```python
# insted of passing column in group by we can also pass function
df = df.set_index('STNAME')
def set_bach_number(item):
  if item[0]<'M':
    return 0
  if item[0]<'Q':
    return 1
  return 2


for group, frame in df.groupby(set_bach_number):
  print(f'group {group} : {len(frame)} records')
```

```python
df = pd.read_csv('listings.csv')
df = df[['cancellation_policy','review_scores_value']]
df.head()
```

```python
def cal_mean_review_scores(group):
  avg = np.nanmean(group['review_scores_value'])

  group['mean_review_score'] = np.abs(avg-group['review_scores_value'])
  return group


df.groupby('cancellation_policy').apply(cal_mean_review_scores).head()
```
